```python
# 학교급구분이 "초등학교"인 데이터만 추출
df_primary = df[df['학교급구분'] == '초등학교']
df_primary
```

```python
# 설립형태가 "공립"인 데이터만 추출
school_type = df['설립형태'] == '공립'
df.loc[school_type, '설립형태'].value_counts()
```

```python
# 소재지가 "제주특별자치도"이면서 학교급구분이 "고등학교"인 데이터만 추출
df_jeju_high = df[
    (df['학교급구분'] == '고등학교') &
    (df['소재지지번주소'].str.contains('제주특별자치도'))
]
df_jeju_high
```

```python
df['설립형태'].value_counts()
```

```python
# 공립 학교가 가장 많음
df['설립형태'].value_counts().plot.pie(figsize=(10, 10))
```

```python
# 소재지가 "제주특별자치도"인 데이터 추출
df_jeju = df[df['소재지지번주소'].str.contains('제주특별자치도')]
df_jeju
```

```python
plt.figure(figsize=(12, 9))
sns.countplot(data=df_jeju, y='학교급구분')
```

```python
# 그리드 그래프에 경도, 위도 기준으로 표기
# 그래프 위쪽 2개 점은 추자도임
df_jeju[['경도', '위도', '학교급구분']].plot.scatter(x='경도', y='위도', figsize=(12, 9), grid=True)
```

```python
# seaborn으로 학교급구분 표기
plt.figure(figsize=(12, 9))
sns.scatterplot(
    data=df_jeju,
    x='경도',
    y='위도',
    hue='학교급구분'
)
```

```python
# seaborn으로 설립형태 표기
plt.figure(figsize=(12, 9))
sns.scatterplot(
    data=df_jeju,
    x='경도',
    y='위도',
    hue='설립형태'
)
```

```python
import folium
```

```python
# folium으로 지도에 마커 표기
Map = folium.Map(location=[df_jeju['위도'].mean(), df_jeju['경도'].mean()], zoom_start=10)

for n in df_jeju.index:
    name = df_jeju.loc[n, '학교명']
    address = df_jeju.loc[n, '소재지도로명주소']
    popup_name = f'{name} - {address}'
    location = [df_jeju.loc[n, '위도'], df_jeju.loc[n, '경도']]
    
    folium.Marker(
        location=location,
        popup=popup_name
    ).add_to(Map)
    
Map
```
